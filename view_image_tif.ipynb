```python
import rasterio
import matplotlib.pyplot as plt

# Open the multi-band TIFF file
with rasterio.open('/data/ai_club/fire/WildfireSpreadLS/2018/fire_21458798/2018-01-02.tif') as src:
    band1 = src.read(1)  # Read the first band
    print(f"Image shape: {band1.shape}")
    
# Display the first band
plt.imshow(band1)
plt.colorbar()
plt.show()
```
